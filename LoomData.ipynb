```python
from tqdm import tqdm 
from pathlib import Path 
main_dict= {}
ngram1 = Path('/home/ajanco/projects/Kat-Muncnik/Jstor_au_Helen_Muchnic/ngram3')
for file in tqdm(ngram1.iterdir(), total=len(list(ngram1.iterdir()))):
    text = file.read_text()
    for line in text.split('\n'):
        word=line.split('\t')[0]
        freq=line.split('\t')[-1]
        if freq == '':
            pass
        else:
            freq = int(freq)
            assert isinstance(freq, int)

        if word in main_dict.keys():
            main_dict[word] = main_dict[word] + freq
        else:
            main_dict[word] = freq
```

```python
names = list(df['eng'])
for key in main_dict.keys():
    for part in key.split():
        for name in names:
            if part.lower() in name:
                print(part)
```
